```python
import numpy as np

# Вектори
a = np.array([1, 1/3, 0])
b = np.array([0, 2, 1/4])
c = np.array([1/2, 1/2, 1])

# Змішаний добуток для об'єму
volume = np.abs(np.dot(a, np.cross(b, c)))

# Векторні добутки для площ
area_ab = np.linalg.norm(np.cross(a, b))
area_ac = np.linalg.norm(np.cross(a, c))
area_bc = np.linalg.norm(np.cross(b, c))

# Повна площа поверхні
total_surface_area = 2 * (area_ab + area_ac + area_bc)

# Кути між ребрами
cos_ab = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
cos_ac = np.dot(a, c) / (np.linalg.norm(a) * np.linalg.norm(c))
cos_bc = np.dot(b, c) / (np.linalg.norm(b) * np.linalg.norm(c))

angle_ab = np.arccos(cos_ab) * 180 / np.pi
angle_ac = np.arccos(cos_ac) * 180 / np.pi
angle_bc = np.arccos(cos_bc) * 180 / np.pi

# Координати інших вершин
M4 = a + b
M5 = a + c
M6 = b + c
M7 = a + b + c

# Вивід результатів
print("Об'єм паралелепіпеда:", volume)
print("Площа повної поверхні паралелепіпеда:", total_surface_area)
print("Кут між векторами a і b:", angle_ab)
print("Кут між векторами a і c:", angle_ac)
print("Кут між векторами b і c:", angle_bc)

print("Координати інших вершин паралелепіпеда:")
print("M4:", M4)
print("M5:", M5)
print("M6:", M6)
print("M7:", M7)
```
