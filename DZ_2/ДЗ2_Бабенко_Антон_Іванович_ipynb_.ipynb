```python
import numpy as np
A = np.array([
    [-1, 1, 2],
    [0, -1, -3],
    [4, -3, 2]
])
B = np.array([
    [1, -4, 7]
])
```

- Напиши функцію для вирішення системи методом Крамера

```python
# Напиши функцію для вирішення системи матричним методом
def solve_inv_matrix(a, b, verbose=False):
    # Перевірка, чи матриця A квадратна
    if A.shape[0] != A.shape[1]:
        return "Матриця A не квадратна"
    try:
      A_inv = np.linalg.inv(A)
    except np.linalg.LinAlgError:
      return "Матриця А не обернена!!!!"

    X = np.dot(A_inv, B)

    if verbose:
        print("Обернена матриця A:\n", A_inv)
        print("Матриця B:\n", B)
        print("Результат множення A_inv на B:\n", X)

    return X

# Приклад використання функції
A = np.array([[-1, 1, 2], [0, -1, -3], [4, -3, 2]])
B = np.array([1, -4, 7])
print(f"Вектор рішення: \r\n {solve_inv_matrix(A, B, verbose=True)}")
```

```python
# Напиши функцію для вирішення системи методом Крамера
def solve_cramer(A, B, verbose=False):
    # Обчислення визначника матриці A
    det_A = np.linalg.det(A)

    if det_A == 0:
        return "Матриця A не має оберненої"

    # Обчислення розв'язків за формулами Крамера
    X = np.zeros(A.shape[1])
    for i in range(A.shape[1]):
        # Створення матриці A_i шляхом заміни i-го стовпця на B
        A_i = A.copy()
        A_i[:, i] = B
        X[i] = np.linalg.det(A_i) / det_A

    if verbose:
        print("Визначник матриці A:", det_A)
        for i in range(A.shape[1]):
            print(f"Матриця A з заміненим {i+1}-м стовпцем:|n", A_i)
            print("Її визначник:", np.linalg.det(A_i))
    return X

# Приклад використання функції
print(f"Вектор рішення: \r\n {solve_cramer(A, B, verbose=True)}")
```
